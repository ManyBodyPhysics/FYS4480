# Coupled Cluster Doubles (CCD) for the Pairing Model

This notebook implements and compares three methods for solving the pairing Hamiltonian:

1. **Exact Diagonalization** - Full numerical solution
2. **Many-Body Perturbation Theory (MBPT2)** - Second-order perturbation theory
3. **Coupled Cluster Doubles (CCD)** - Non-perturbative coupled cluster method

## The Pairing Hamiltonian

The pairing Hamiltonian describes fermions in doubly degenerate single-particle levels with pairing interactions:

$$H = \sum_i \epsilon_i n_i - \frac{g}{2} \sum_{ij} P_i^\dagger P_j$$

where $P_i^\dagger = a_{i\uparrow}^\dagger a_{i\downarrow}^\dagger$ creates a pair in level $i$.

```python
# Imports
from functools import partial
import jax
import jax.numpy as jnp
import numpy as np
from jax import Array, jit
import matplotlib.pyplot as plt
from itertools import combinations
from scipy.linalg import eigh

jax.config.update("jax_enable_x64", True)

print("Imports complete")
```

## 1. Exact Diagonalization

We first implement exact diagonalization of the pairing Hamiltonian in the seniority-zero subspace (no broken pairs).

```python
def construct_pairing_hamiltonian_exact(N, n_levels, g, delta):
    """
    Construct pairing Hamiltonian in seniority-zero basis.
    
    Args:
        N: Number of fermions (must be even)
        n_levels: Total number of doubly degenerate levels
        g: Pairing strength
        delta: Single-particle level spacing
    
    Returns:
        H: Hamiltonian matrix
        basis_states: List of basis states (tuples of occupied levels)
    """
    if N % 2 != 0:
        raise ValueError("N must be even (seniority v=0)")
    
    n_pairs = N // 2
    
    # Generate basis: all ways to place n_pairs among n_levels
    basis_states = list(combinations(range(n_levels), n_pairs))
    n_basis = len(basis_states)
    
    # Single-particle energies
    epsilon_levels = 0.5 * delta * np.arange(n_levels)
    
    H = np.zeros((n_basis, n_basis), dtype=float)
    
    # Construct Hamiltonian matrix elements
    for i, state_i in enumerate(basis_states):
        for j, state_j in enumerate(basis_states):
            
            # Diagonal: single-particle energies
            if i == j:
                # Each level contributes 2*epsilon (spin up and down)
                H[i, j] = sum(2 * epsilon_levels[k] for k in state_i)
            
            # Off-diagonal: pairing interaction
            # <i|P_k† P_l|j> where P_k† creates pair in level k
            for k in range(n_levels):
                for l in range(n_levels):
                    # Check if we can scatter pair from level l to level k
                    if l in state_j and k not in state_j:
                        # Remove l, add k
                        state_temp = list(state_j)
                        state_temp.remove(l)
                        state_temp.append(k)
                        state_temp = tuple(sorted(state_temp))
                        
                        if state_temp == state_i:
                            H[i, j] -= g / 2.0
    
    return H, basis_states


def exact_diagonalization(pnum, hnum, delta, g):
    """
    Perform exact diagonalization for the pairing model.
    
    Args:
        pnum: Number of particle states
        hnum: Number of hole states  
        delta: Single-particle spacing
        g: Pairing strength
    
    Returns:
        E_ground: Ground state energy
        eigenvalues: All eigenvalues
        eigenvectors: All eigenvectors
        basis_states: Basis states
    """
    N = 2 * hnum  # Number of fermions (2 per hole level)
    n_levels = hnum + pnum  # Total number of levels
    
    H, basis_states = construct_pairing_hamiltonian_exact(N, n_levels, g, delta)
    
    # Diagonalize
    eigenvalues, eigenvectors = eigh(H)
    
    E_ground = eigenvalues[0]
    
    return E_ground, eigenvalues, eigenvectors, basis_states


print("Exact diagonalization functions defined")
```

## 2. CCD Implementation (from JAX code)

The Coupled Cluster Doubles method uses the exponential ansatz:

$$|\Psi\rangle = e^T |\Phi_0\rangle$$

where $T = T_2$ contains only double excitations (pair scattering).

```python
def init_pairing_v(g: float, pnum: int, hnum: int) -> tuple[Array, Array, Array]:
    """Returns potential matrices of the pairing model in three relevant channels.

    Args:
        g (float): Strength of the pairing interaction.
        pnum (int): Number of particle states.
        hnum (int): Number of hole states.

    Returns:
        tuple: A tuple containing three numpy arrays:
        The interaction as a 4-indexed tensor in the:
        - v_pppp (Array): particle-particle channel (pnum, pnum, pnum, pnum).
        - v_pphh (Array): particle-hole channel (pnum, pnum, hnum, hnum).
        - v_hhhh (Array): hole-hole channel (hnum, hnum, hnum, hnum).
    """

    v_pppp = np.zeros((pnum, pnum, pnum, pnum))
    v_pphh = np.zeros((pnum, pnum, hnum, hnum))
    v_hhhh = np.zeros((hnum, hnum, hnum, hnum))

    gval = -0.5 * g

    # Even indices for particles and holes
    p_even = np.arange(0, pnum, 2)
    h_even = np.arange(0, hnum, 2)

    # For v_pppp
    for a in p_even:
        for b in p_even:
            v_pppp[a, a + 1, b, b + 1] = gval
            v_pppp[a + 1, a, b, b + 1] = -gval
            v_pppp[a, a + 1, b + 1, b] = -gval
            v_pppp[a + 1, a, b + 1, b] = gval

    # For v_pphh
    for a in p_even:
        for i in h_even:
            v_pphh[a, a + 1, i, i + 1] = gval
            v_pphh[a + 1, a, i, i + 1] = -gval
            v_pphh[a, a + 1, i + 1, i] = -gval
            v_pphh[a + 1, a, i + 1, i] = gval

    # For v_hhhh
    for j in h_even:
        for i in h_even:
            v_hhhh[j, j + 1, i, i + 1] = gval
            v_hhhh[j + 1, j, i, i + 1] = -gval
            v_hhhh[j, j + 1, i + 1, i] = -gval
            v_hhhh[j + 1, j, i + 1, i] = gval
            
    v_pppp, v_pphh, v_hhhh = map(jnp.array, (v_pppp, v_pphh, v_hhhh))

    return v_pppp, v_pphh, v_hhhh


def init_pairing_fock(
    delta: float, g: float, pnum: int, hnum: int
) -> tuple[Array, Array]:
    """Initialize the Fock matrix of the pairing model.

    Args:
        delta (float): Single-particle spacing.
        g (float): Pairing strength.
        pnum (int): Number of particle states.
        hnum (int): Number of hole states.

    Returns:
        tuple: A tuple containing two numpy arrays:
        The Fock matrix in two channels (pnum, pnum), (hnum, hnum).
    """
    deltaval = 0.5 * delta
    gval = -0.5 * g
    f_pp = np.zeros((pnum, pnum))
    f_hh = np.zeros((hnum, hnum))

    # Set f_hh diagonals
    i_even = np.arange(0, hnum, 2)
    f_hh[i_even, i_even] = deltaval * i_even + gval
    f_hh[i_even + 1, i_even + 1] = deltaval * i_even + gval

    # Set f_pp diagonals
    a_even = np.arange(0, pnum, 2)
    f_pp[a_even, a_even] = deltaval * (hnum + a_even)
    f_pp[a_even + 1, a_even + 1] = deltaval * (hnum + a_even)

    f_pp, f_hh = jnp.array(f_pp), jnp.array(f_hh)
    return f_pp, f_hh


def init_t2(v_pphh: Array, f_pp: Array, f_hh: Array) -> Array:
    """Initialize T2 amplitudes as in MBPT2.

    Args:
        v_pphh (Array): Particle-hole interaction tensor.
        f_pp (Array): Fock matrix for particle states.
        f_hh (Array): Fock matrix for hole states.

    Returns:
        Array: Initial T2 amplitudes.
    """
    Fhh_diag = jnp.diag(f_hh)
    Fpp_diag = jnp.diag(f_pp)

    # denominator shape: (pnum, pnum, hnum, hnum)
    denom = (
        Fhh_diag[None, None, :, None]
        + Fhh_diag[None, None, None, :]
        - Fpp_diag[:, None, None, None]
        - Fpp_diag[None, :, None, None]
    )

    t2_new = v_pphh / denom
    return t2_new


@jit
def ccd_iter(
    v_pppp: Array, v_pphh: Array, v_hhhh: Array, f_pp: Array, f_hh: Array, t2: Array
) -> Array:
    """Perform one iteration of the CCD equations.

    Args:
        v_pppp (Array): Pairing tensor in pppp channel.
        v_pphh (Array): Pairing tensor in pphh channel.
        v_hhhh (Array): Pairing tensor in hhhh channel.
        f_pp (Array): Fock matrix in pp channel.
        f_hh (Array): Fock matrix in hh channel.
        t2 (Array): Initial T2 amplitude.

    Returns:
        Array: New T2 amplitude.
    """
    Hbar_pphh = (
        v_pphh
        + jnp.einsum("bc,acij->abij", f_pp, t2)
        - jnp.einsum("ac,bcij->abij", f_pp, t2)
        - jnp.einsum("abik,kj->abij", t2, f_hh)
        + jnp.einsum("abjk,ki->abij", t2, f_hh)
        + 0.5 * jnp.einsum("abcd,cdij->abij", v_pppp, t2)
        + 0.5 * jnp.einsum("abkl,klij->abij", t2, v_hhhh)
    )

    # hh intermediate
    chi_hh = 0.5 * jnp.einsum("cdkl,cdjl->kj", v_pphh, t2)

    Hbar_pphh = Hbar_pphh - (
        jnp.einsum("abik,kj->abij", t2, chi_hh) - jnp.einsum("abik,kj->abji", t2, chi_hh)
    )

    # pp intermediate
    chi_pp = -0.5 * jnp.einsum("cdkl,bdkl->cb", v_pphh, t2)

    Hbar_pphh = Hbar_pphh + (
        jnp.einsum("acij,cb->abij", t2, chi_pp) - jnp.einsum("acij,cb->baij", t2, chi_pp)
    )

    # hhhh intermediate
    chi_hhhh = 0.5 * jnp.einsum("cdkl,cdij->klij", v_pphh, t2)

    Hbar_pphh = Hbar_pphh + 0.5 * jnp.einsum("abkl,klij->abij", t2, chi_hhhh)

    # phph intermediate
    chi_phph = +0.5 * jnp.einsum("cdkl,dblj->bkcj", v_pphh, t2)

    Hbar_pphh = Hbar_pphh + (
        jnp.einsum("bkcj,acik->abij", chi_phph, t2)
        - jnp.einsum("bkcj,acik->baij", chi_phph, t2)
        - jnp.einsum("bkcj,acik->abji", chi_phph, t2)
        + jnp.einsum("bkcj,acik->baji", chi_phph, t2)
    )

    Fhh_diag = jnp.diag(f_hh)
    Fpp_diag = jnp.diag(f_pp)
    denom = (
        Fhh_diag[None, None, :, None]
        + Fhh_diag[None, None, None, :]
        - Fpp_diag[:, None, None, None]
        - Fpp_diag[None, :, None, None]
    )

    t2_new = t2 + Hbar_pphh / denom
    return t2_new


@jit
def ccd_energy(v_pphh: Array, t2: Array) -> Array:
    """Compute CCD energy.

    Args:
        v_pphh (Array): Pairing tensor in pphh channel.
        t2 (Array): T2 amplitude.

    Returns:
        Array: CCD energy.
    """
    return 0.25 * jnp.sum(v_pphh * t2)


def calculate_ccd_energy(
    pnum: int, hnum: int, delta: float, g: float, verbose: bool = False
) -> tuple[float, float]:
    """Calculate the CCD and MBPT2 energies for the pairing model.

    Args:
        pnum (int): Number of particle states.
        hnum (int): Number of hole states.
        delta (float): Single-particle spacing.
        g (float): Pairing strength.
        verbose (bool): Whether to print the results.

    Returns:
        tuple: (CCD energy, MBPT2 energy)
    """
    if verbose:
        print("Parameters:")
        print(f"{delta = }, {g = }, {pnum = }, {hnum = }")

    # Initialize pairing matrix elements and Fock matrix
    v_pppp, v_pphh, v_hhhh = init_pairing_v(g, pnum, hnum)
    f_pp, f_hh = init_pairing_fock(delta, g, pnum, hnum)

    # Initialize T2 amplitudes from MBPT2
    t2 = init_t2(v_pphh, f_pp, f_hh)
    mbpt2_erg = ccd_energy(v_pphh, t2)

    # iterate CCD equations niter times
    niter = 200
    mix = 0.3
    erg_old = 0.0
    eps = jnp.float64(1.0e-10)
    for iter_count in range(niter):
        t2_new = ccd_iter(v_pppp, v_pphh, v_hhhh, f_pp, f_hh, t2)
        erg = ccd_energy(v_pphh, t2_new)
        myeps = jnp.abs(erg - erg_old) / jnp.abs(erg)
        if myeps < eps:
            t2 = t2_new
            num_iter = iter_count
            break
        erg_old = erg
        t2 = mix * t2_new + (1.0 - mix) * t2

    if verbose:
        print(f"MBPT2 energy = {mbpt2_erg}")
        print(f"Converged in {str(num_iter):3s} iterations, CCD Energy = {erg}")
        
    return float(erg), float(mbpt2_erg)


print("CCD functions defined")
```

## 3. Single Point Calculation

Let's test all three methods for a single value of the pairing strength $g$.

```python
# Set parameters
pnum = 4  # number of particle states
hnum = 4  # number of hole states
delta = 1.0  # level spacing
g_test = 0.5  # pairing strength

print("="*80)
print("SINGLE POINT COMPARISON")
print("="*80)
print(f"\nParameters: pnum={pnum}, hnum={hnum}, delta={delta}, g={g_test}")

# Exact diagonalization
E_exact, eigenvalues, eigenvectors, basis_states = exact_diagonalization(pnum, hnum, delta, g_test)

print(f"\n{'='*80}")
print("EXACT DIAGONALIZATION")
print("="*80)
print(f"\nGround state energy: {E_exact:.10f}")
print(f"\nAll eigenvalues:")
for i, E in enumerate(eigenvalues[:10]):
    print(f"  E_{i}: {E:+.10f}")

print(f"\nGround state composition:")
psi_exact = eigenvectors[:, 0]
for i, coeff in enumerate(psi_exact):
    if abs(coeff) > 0.01:
        print(f"  {abs(coeff)**2:.4f} |{basis_states[i]}⟩: {coeff:+.6f}")

# CCD and MBPT2
E_ccd, E_mbpt2 = calculate_ccd_energy(pnum, hnum, delta, g_test, verbose=False)

# The reference energy (HF energy for holes)
E_ref = 2 - g_test  # Reference state energy

# Total energies
E_ccd_total = E_ref + E_ccd
E_mbpt2_total = E_ref + E_mbpt2

print(f"\n{'='*80}")
print("MBPT2")
print("="*80)
print(f"Correlation energy: {E_mbpt2:.10f}")
print(f"Total energy:       {E_mbpt2_total:.10f}")
print(f"Error vs exact:     {abs(E_mbpt2_total - E_exact):.6e}")

print(f"\n{'='*80}")
print("CCD")
print("="*80)
print(f"Correlation energy: {E_ccd:.10f}")
print(f"Total energy:       {E_ccd_total:.10f}")
print(f"Error vs exact:     {abs(E_ccd_total - E_exact):.6e}")

print(f"\n{'='*80}")
print("SUMMARY")
print("="*80)
print(f"\n{'Method':20s} | {'Total Energy':>14s} | {'Error':>12s}")
print("-"*55)
print(f"{'Exact':20s} | {E_exact:+14.10f} | {'0.00e+00':>12s}")
print(f"{'MBPT2':20s} | {E_mbpt2_total:+14.10f} | {abs(E_mbpt2_total-E_exact):12.2e}")
print(f"{'CCD':20s} | {E_ccd_total:+14.10f} | {abs(E_ccd_total-E_exact):12.2e}")
```

## 4. Energy Scan vs Pairing Strength

Now let's compute the ground state energy as a function of the pairing strength $g$ using all three methods.

```python
# Scan over g values
g_values = np.linspace(-1, 1, 50)

energies_exact = np.zeros_like(g_values)
energies_ccd = np.zeros_like(g_values)
energies_mbpt2 = np.zeros_like(g_values)

print("Computing energies for different g values...")
for i, g in enumerate(g_values):
    # Exact
    E_exact, _, _, _ = exact_diagonalization(pnum, hnum, delta, g)
    energies_exact[i] = E_exact
    
    # CCD and MBPT2
    E_ccd, E_mbpt2 = calculate_ccd_energy(pnum, hnum, delta, g, verbose=False)
    E_ref = 2 - g
    energies_ccd[i] = E_ref + E_ccd
    energies_mbpt2[i] = E_ref + E_mbpt2
    
    if (i+1) % 10 == 0:
        print(f"  Progress: {i+1}/{len(g_values)}")

print("\n✓ Energy scan complete")
```

## 5. Visualization and Comparison

Create comprehensive plots comparing all three methods.

```python
fig, axes = plt.subplots(2, 2, figsize=(14, 10))

# Plot 1: Total energies
ax1 = axes[0, 0]
ax1.plot(g_values, energies_exact, 'k-', linewidth=2.5, label='Exact', zorder=3)
ax1.plot(g_values, energies_ccd, 'b--', linewidth=2, label='CCD', zorder=2)
ax1.plot(g_values, energies_mbpt2, 'r:', linewidth=2, label='MBPT2', zorder=1)
ax1.set_xlabel('Pairing strength g', fontsize=12)
ax1.set_ylabel('Ground state energy', fontsize=12)
ax1.set_title('Energy vs Pairing Strength', fontweight='bold', fontsize=13)
ax1.legend(fontsize=11, loc='best')
ax1.grid(True, alpha=0.3)

# Plot 2: Absolute errors
ax2 = axes[0, 1]
error_ccd = np.abs(energies_ccd - energies_exact)
error_mbpt2 = np.abs(energies_mbpt2 - energies_exact)
ax2.semilogy(g_values, error_mbpt2, 'r-', linewidth=2, label='MBPT2', marker='o', markersize=3)
ax2.semilogy(g_values, error_ccd, 'b-', linewidth=2, label='CCD', marker='s', markersize=3)
ax2.set_xlabel('Pairing strength g', fontsize=12)
ax2.set_ylabel('|E - E_exact|', fontsize=12)
ax2.set_title('Absolute Error (log scale)', fontweight='bold', fontsize=13)
ax2.legend(fontsize=11)
ax2.grid(True, alpha=0.3, which='both')

# Plot 3: Relative errors
ax3 = axes[1, 0]
# Avoid division by zero
rel_error_ccd = np.abs((energies_ccd - energies_exact) / (energies_exact + 1e-10)) * 100
rel_error_mbpt2 = np.abs((energies_mbpt2 - energies_exact) / (energies_exact + 1e-10)) * 100
ax3.plot(g_values, rel_error_mbpt2, 'r-', linewidth=2, label='MBPT2', marker='o', markersize=3)
ax3.plot(g_values, rel_error_ccd, 'b-', linewidth=2, label='CCD', marker='s', markersize=3)
ax3.set_xlabel('Pairing strength g', fontsize=12)
ax3.set_ylabel('Relative error (%)', fontsize=12)
ax3.set_title('Relative Error', fontweight='bold', fontsize=13)
ax3.legend(fontsize=11)
ax3.grid(True, alpha=0.3)

# Plot 4: Energy differences (zoomed)
ax4 = axes[1, 1]
ax4.plot(g_values, energies_exact - energies_mbpt2, 'r-', linewidth=2, 
         label='Exact - MBPT2', marker='o', markersize=3)
ax4.plot(g_values, energies_exact - energies_ccd, 'b-', linewidth=2, 
         label='Exact - CCD', marker='s', markersize=3)
ax4.axhline(0, color='k', linestyle='--', linewidth=1, alpha=0.5)
ax4.set_xlabel('Pairing strength g', fontsize=12)
ax4.set_ylabel('Energy difference', fontsize=12)
ax4.set_title('Exact - Approximate Methods', fontweight='bold', fontsize=13)
ax4.legend(fontsize=11)
ax4.grid(True, alpha=0.3)

plt.suptitle(f'Pairing Model Comparison (pnum={pnum}, hnum={hnum}, δ={delta})', 
             fontsize=14, fontweight='bold', y=0.995)
plt.tight_layout()
plt.savefig('pairing_model_comparison.png', dpi=150, bbox_inches='tight')
plt.show()

print("\n✓ Plots saved as 'pairing_model_comparison.png'")
```

## 6. Detailed Analysis at Specific Points

Let's examine the accuracy at several representative values of $g$.

```python
# Analyze at specific g values
g_test_points = [-0.5, 0.0, 0.5, 1.0]

print("="*80)
print("DETAILED ANALYSIS AT SPECIFIC g VALUES")
print("="*80)

for g in g_test_points:
    print(f"\n{'='*80}")
    print(f"g = {g}")
    print(f"{'='*80}")
    
    # Exact
    E_exact, eigenvalues, eigenvectors, basis_states = exact_diagonalization(pnum, hnum, delta, g)
    
    # CCD and MBPT2
    E_ccd, E_mbpt2 = calculate_ccd_energy(pnum, hnum, delta, g, verbose=False)
    E_ref = 2 - g
    E_ccd_total = E_ref + E_ccd
    E_mbpt2_total = E_ref + E_mbpt2
    
    print(f"\n{'Method':15s} | {'Total Energy':>14s} | {'Abs Error':>12s} | {'Rel Error (%)':>15s}")
    print("-"*70)
    print(f"{'Exact':15s} | {E_exact:+14.10f} | {'0.00e+00':>12s} | {'0.00':>15s}")
    
    abs_err_mbpt2 = abs(E_mbpt2_total - E_exact)
    rel_err_mbpt2 = abs_err_mbpt2 / abs(E_exact) * 100 if E_exact != 0 else 0
    print(f"{'MBPT2':15s} | {E_mbpt2_total:+14.10f} | {abs_err_mbpt2:12.2e} | {rel_err_mbpt2:15.6f}")
    
    abs_err_ccd = abs(E_ccd_total - E_exact)
    rel_err_ccd = abs_err_ccd / abs(E_exact) * 100 if E_exact != 0 else 0
    print(f"{'CCD':15s} | {E_ccd_total:+14.10f} | {abs_err_ccd:12.2e} | {rel_err_ccd:15.6f}")
    
    # Spectrum
    print(f"\nLowest 5 eigenvalues:")
    for i in range(min(5, len(eigenvalues))):
        print(f"  E_{i}: {eigenvalues[i]:+.10f}")
```

## 7. Summary Statistics

Compute overall statistics across all $g$ values.

```python
# Calculate statistics
max_error_mbpt2 = np.max(np.abs(energies_mbpt2 - energies_exact))
mean_error_mbpt2 = np.mean(np.abs(energies_mbpt2 - energies_exact))
rms_error_mbpt2 = np.sqrt(np.mean((energies_mbpt2 - energies_exact)**2))

max_error_ccd = np.max(np.abs(energies_ccd - energies_exact))
mean_error_ccd = np.mean(np.abs(energies_ccd - energies_exact))
rms_error_ccd = np.sqrt(np.mean((energies_ccd - energies_exact)**2))

print("="*80)
print("OVERALL STATISTICS")
print("="*80)
print(f"\nAcross {len(g_values)} values of g from {g_values[0]:.2f} to {g_values[-1]:.2f}:")
print(f"\n{'Statistic':20s} | {'MBPT2':>15s} | {'CCD':>15s}")
print("-"*60)
print(f"{'Maximum error':20s} | {max_error_mbpt2:15.6e} | {max_error_ccd:15.6e}")
print(f"{'Mean absolute error':20s} | {mean_error_mbpt2:15.6e} | {mean_error_ccd:15.6e}")
print(f"{'RMS error':20s} | {rms_error_mbpt2:15.6e} | {rms_error_ccd:15.6e}")

print(f"\n{'='*80}")
print("CONCLUSIONS")
print("="*80)

improvement_factor = mean_error_mbpt2 / mean_error_ccd if mean_error_ccd > 0 else float('inf')

print(f"\n1. CCD is {improvement_factor:.1f}× more accurate than MBPT2 (mean absolute error)")
print(f"\n2. CCD captures non-perturbative effects beyond MBPT2")
print(f"\n3. For this pairing model:")
print(f"   - MBPT2 accuracy: ~{mean_error_mbpt2:.2e} (mean absolute error)")
print(f"   - CCD accuracy:   ~{mean_error_ccd:.2e} (mean absolute error)")
print(f"\n4. Both methods converge to exact result in appropriate limits")
print(f"\n5. CCD provides near-exact results across the entire range of g")
```

## 8. Additional Visualization: Error Ratio

Show how much better CCD performs compared to MBPT2.

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))

# Plot 1: Error ratio
# Avoid division by very small numbers
error_ratio = np.where(error_ccd > 1e-12, error_mbpt2 / error_ccd, 1.0)
ax1.plot(g_values, error_ratio, 'g-', linewidth=2.5, marker='o', markersize=4)
ax1.axhline(1, color='k', linestyle='--', linewidth=1, alpha=0.5, label='Equal accuracy')
ax1.set_xlabel('Pairing strength g', fontsize=12)
ax1.set_ylabel('Error(MBPT2) / Error(CCD)', fontsize=12)
ax1.set_title('CCD Improvement Factor over MBPT2', fontweight='bold', fontsize=13)
ax1.legend(fontsize=11)
ax1.grid(True, alpha=0.3)
ax1.set_ylim(bottom=0)

# Plot 2: Both errors on same plot (linear scale, small g)
mask = np.abs(g_values) < 0.6  # Focus on small g region
ax2.plot(g_values[mask], error_mbpt2[mask], 'r-', linewidth=2, label='MBPT2', marker='o', markersize=4)
ax2.plot(g_values[mask], error_ccd[mask], 'b-', linewidth=2, label='CCD', marker='s', markersize=4)
ax2.set_xlabel('Pairing strength g', fontsize=12)
ax2.set_ylabel('Absolute error', fontsize=12)
ax2.set_title('Error Comparison (|g| < 0.6)', fontweight='bold', fontsize=13)
ax2.legend(fontsize=11)
ax2.grid(True, alpha=0.3)

plt.tight_layout()
plt.savefig('pairing_model_error_comparison.png', dpi=150, bbox_inches='tight')
plt.show()

print("\n✓ Additional plots saved as 'pairing_model_error_comparison.png'")
```

## Summary

This notebook demonstrates that:

1. **Exact Diagonalization** provides the benchmark ground truth
2. **MBPT2** gives reasonable first-order corrections but has limitations
3. **CCD** captures non-perturbative correlation effects and achieves near-exact accuracy

The coupled cluster method systematically improves upon perturbation theory by including higher-order correlations through the exponential ansatz $e^T|\Phi_0\rangle$, making it one of the most accurate methods in quantum chemistry and nuclear structure theory.
